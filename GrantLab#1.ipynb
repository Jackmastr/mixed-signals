```python
%matplotlib inline
import matplotlib.pylab as pl
import numpy as np
import ugradio
i=1
while i<10:
    filename = "sample_" + str(i) + ".npy"
    arr_plot = np.load(filename)
    pl.figure(i, figsize=(20,10))
    pl.plot (arr_plot[:256], 'k.-')
    pl.xlabel ('Time')
    pl.ylabel ('Voltage')
    i += 1
```

```python
i=1
while i<10:
    filename = "sample_" + str(i) + ".npy"
    test_arr = np.load(filename)
    fourier_plot = ugradio.dft.dft(test_arr[:256], 128/6.25)
    fourier_plot = np.abs(fourier_plot)**2
    pl.xlabel('Frequency')
    pl.ylabel('Amplitude')
    pl.figure(i, figsize=(20,10))
    pl.plot(fourier_plot)
    i += 1
```
